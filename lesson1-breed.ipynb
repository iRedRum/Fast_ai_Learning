# 3 Initial model

if you get any error information like:CUDA out of memory, just restart Kernel, and decrease the batch size.

```python
def get_data(sz,bs):
    tfms = tfms_from_model(arch, sz, aug_tfms=transforms_side_on, max_zoom=1.1)
    data=ImageClassifierData.from_csv(PATH,'train',f'{PATH}labels.csv',test_name='test',
                                  val_idxs=val_idxs,suffix='.jpg',tfms=tfms,bs=bs)
    #data=ImageClassifierData.from_csv(PATH,'train',f'{PATH}labels.csv',test_name='test',
     #                             suffix='.jpg',tfms=tfms,bs=bs)
    return data if sz>300 else data.resize(340,'tmp')
```

## 3.1 Precompute

```python
data=get_data(sz,bs)
```

```python
data.sz
```

Here we set our Dropout = 50% to avoid Overfitting.

```python
learn=ConvLearner.pretrained(arch,data,precompute=True)
```

# Increase size

We firstly start with training our model with small size of image, and then we resize the images to bigger ones.
This trick can help us to avoid overfitting, because you are using different size of images.
Well, conceptually, you are still using the same datasets to train the model, so it is a different insight to train your model.

```python
learn.set_data(get_data(400,bs))
learn.freeze()
```

```python
learn.fit(1e-2,2,cycle_len=1)
```

```python
learn.sched.plot_lr()
```

```python
learn.fit(1e-2,3,cycle_len=1,cycle_mult=2)
```

```python
learn.precompute
```

```python
learn.sched.plot_lr()
```

```python
log_preds,y=learn.TTA()
```

seems that what i do below is the correct way to figure out our accuracy after TTA

```python
probs = np.mean(np.exp(log_preds),0)
accuracy_np(probs, y)
```

# Create Submission

```python
log_preds,y=learn.TTA(is_test=True)
```

```python
probs = np.mean(np.exp(log_preds),0)
```

```python
probs.shape
```

```python
df=pd.DataFrame(probs)
df.columns=data.classes
```

```python
df.insert(0,'id',[o[5:-4] for o in data.test_ds.fnames])
```
